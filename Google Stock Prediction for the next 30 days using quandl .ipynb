```python
import quandl
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# To get the stock data
df=quandl.get("WIKI/GOOG")
# To print the data for grtting it's view
print(df.head())
```

```python
#To get the adjusted close price
df=df[['Adj. Close']]
# To get a look at the new data
print(df.head())
```

```python
# A variable to predict the n number of days in future
forecast=30
# A new column is been created that has been shifted up n times
df['Prediction'] = df[['Adj. Close']].shift(-forecast)
# To print the new data
print(df.tail())
```

```python
# Creating an independent data set x and converting it into a numpy array
x = np.array(df.drop(['Prediction'],1))

#To remove the last n rows


x = x[:-forecast]
print(x)
```

```python
# To create a dependent data set y and convert it to a numpy array
y = np.array(df['Prediction'])

# To retrieve all the values of y except the last n rows
y = y[:-forecast]
print(y)
```

```python
# For Spliting the data into 80% training and 20% testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# To create and train the Support Vector Machine (Regressor)
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
svr_rbf.fit(x_train, y_train)
```

```python
# Testing Model: Score returns the coefficient of determination R^2 of the prediction. 
# The best possible score would be 1.0
svm_confidence = svr_rbf.score(x_test, y_test)
print("svm confidence: ", svm_confidence)
```

```python
# To create and train the Linear Regression  Model
lr = LinearRegression()

# To train the model
lr.fit(x_train, y_train)
```

```python
# Testing Model: Score returns the coefficient of determination R^2 of the prediction. 
# The best possible score would be 1.0
lr_confidence = lr.score(x_test, y_test)
print("lr confidence: ", lr_confidence)
```

```python
# Set x_forecast equal to the last 30 rows of the original data set from Adj. Close column
x_forecast = np.array(df.drop(['Prediction'],1))[-forecast:]
print(x_forecast)
```

```python
# To print linear regression model predictions for the next n days
lr_prediction = lr.predict(x_forecast)
print(lr_prediction)

# To print support vector regressor model predictions for the next n days
svm_prediction = svr_rbf.predict(x_forecast)
print(svm_prediction)
```
